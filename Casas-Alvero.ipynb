# Conjectura de Casas-Alvero
A aquest document s'escriurà el codi necessari per verificar que la conjectura és certa per certs graus. Els càlculs es duran a terme mitjançant l'algorisme dels resultants enunciat al cos del treball i les optimitzacions enunciades a la secció de la conjectura.

```python
var('a0 a1 a2 a3 a4 a5 a6')
```

## Grau n=3

```python
P = a1*x + x^3
P1 = diff(P, x); show('P1 = ', P1)
P2 = diff(P, x, 2); show('P2 = ', P2)

print("NIVELL 1")
R_1 = P.resultant(P1, x).factor(); show('  Resultant 1.1 = ', R_1)

# resolem l'equació donada
sol1 = solve(R_1 == 0, a1)
a1_sol = sol1[0].rhs()
print('Resultat del nivell 1: ', sol1[0])

show('El polinomi factoritzat és:    ', P.subs(a1 = a1_sol).factor())
```

## Grau n=4

```python
P= a1*x + a2*x^2 + x^4
P1 = diff(P, x); show('P1 = ', P1)
P2 = diff(P, x, 2); show('P2 = ', P2)
P3 = diff(P, x, 3); show('P3 = ', P3)

print("NIVELL 1")
R_1 = P.resultant(P1, x).factor(); show('  Resultant 1.1 = ', R_1.expand())
R_2 = P.resultant(P2, x).factor(); show('  Resultant 1.2 = ', R_2.expand())

llistaResultants = [R_1, R_2]
print("    ", [f'deg(Resultant 1.{i+1}, a1) = {llistaResultants[i].degree(a1)}' for i in range(len(llistaResultants))])
print("    ", [f'deg(Resultant 1.{i+1}, a2) = {llistaResultants[i].degree(a2)}' for i in range(len(llistaResultants))])
```

```python
# Veiem que la matriu de Sylvester més petita possible és la que fa servir a1 com a variable. 
# Aquesta és de mida 6x6, i l'altra possible és de mida 7x7.
print("NIVELL 2")
R2_1 = R_1.resultant(R_2, a1).factor(); show('  Resultant 2.1 = ', R2_1)
```

```python
# resolem l'equació donada
sol2 = solve(R2_1 == 0, a2)
a2_sol = sol2[0].rhs()
print('Resultat del nivell 2: ', sol2[0])

sol1 = solve([R_1.subs(a2 = a2_sol) == 0, R_2.subs(a2 = a2_sol) == 0], a1)
a1_sol = sol1[0][0].rhs()
print('Resultat del nivell 1: ', sol1[0][0])

show('El polinomi factoritzat és:    ', P.subs(a1 = a1_sol, a2 = a2_sol).factor())
```
